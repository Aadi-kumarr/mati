```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

print("Libraries Loaded")
```

```python
train_df = pd.read_csv(
    "../data/processed/train_engineered.csv"
)

print("Shape:", train_df.shape)

train_df.head()
```

```python
TOP_SENSORS = [
    "sensor_11",
    "sensor_14",
    "sensor_4",
    "sensor_17",
    "sensor_9"
]

TOP_SENSORS
```

```python
for sensor in TOP_SENSORS:

    train_df[f"{sensor}_rolling_std"] = (
        train_df.groupby("unit_number")[sensor]
        .transform(
            lambda x: x.rolling(
                window=10,
                min_periods=1
            ).std()
        )
    )

print("Rolling Std Features Added")
```

```python
for sensor in TOP_SENSORS:

    train_df[f"{sensor}_rolling_min"] = (
        train_df.groupby("unit_number")[sensor]
        .transform(
            lambda x: x.rolling(
                window=10,
                min_periods=1
            ).min()
        )
    )

print("Rolling Min Features Added")
```

```python
for sensor in TOP_SENSORS:

    train_df[f"{sensor}_rolling_max"] = (
        train_df.groupby("unit_number")[sensor]
        .transform(
            lambda x: x.rolling(
                window=10,
                min_periods=1
            ).max()
        )
    )

print("Rolling Max Features Added")
```

```python
for sensor in TOP_SENSORS:

    first_diff = (
        train_df.groupby("unit_number")[sensor]
        .diff()
    )

    second_diff = (
        first_diff.groupby(
            train_df["unit_number"]
        ).diff()
    )

    train_df[
        f"{sensor}_acceleration"
    ] = second_diff

print("Acceleration Features Added")
```

```python
train_df.fillna(
    0,
    inplace=True
)

print(
    train_df.isnull().sum().sum()
)
```

```python
new_features = [
    col
    for col in train_df.columns
    if (
        "rolling" in col
        or "acceleration" in col
    )
]

corrs = (
    train_df[new_features + ["RUL"]]
    .corr()["RUL"]
    .abs()
    .sort_values(
        ascending=False
    )
)

corrs.head(20)
```

```python
train_df.to_csv(
    "../data/processed/train_advanced_features.csv",
    index=False
)

print(
    "Advanced Features Saved Successfully"
)
```
